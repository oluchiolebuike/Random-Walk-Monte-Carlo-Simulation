```python
import random
import matplotlib.pyplot as plt

def random_walk(n):
    """ return coordinates after 'n' block randon walk. """
    x = 0
    y = 0

    for i in range(n):
        # change x or y value based off of direction (N, S, E, W)
        (dx, dy) = random.choice([(0,1), (0, -1), (1, 0),(-1, 0)])
        x += dx
        y += dy

    return (x, y)

number_of_walks = 20000 # if graph looks noisy increase to 50 000 or even 100 000
walk_lengths = list(range(1, 31))
probabilities = []

   
for walk_length in walk_lengths:
    # number of walks 4 or fewer from home
    no_transport = 0 
    
    # monte carlo simulation of 20 000 samples
    for i in range(number_of_walks):
        (x, y) = random_walk(walk_length)
        distance = abs(x) + abs(y)

        if distance <= 4:
            no_transport += 1

    probability = no_transport / number_of_walks
    probabilities.append(probability)

# plot random walk monte carlo simulation graph
plt.figure()
plt.plot(walk_lengths, probabilities, marker='o')

plt.xlabel("Walk length (Number of steps)")
plt.ylabel("Probability of Distance ≤ 4")
plt.title("Random Walk Monte Carlo Simulation")

plt.grid()
plt.show()




```
